```
import numpy as np
import pandas as pd
data = 'Malonylation.elm'
sequence = []
with open(data, 'r', encoding="utf-8") as f:
    for line in f:
        sequence.append(line.split())
```

```
df = pd.DataFrame(sequence)
df.head(10)
df = df.iloc[1: , [2, 4]]
position = df.loc[0:,[2]]
seq = df.loc[0:,[4]]
#position.head()
```

```
#Position Select
pos=position.get_values()
pos=pos.astype(int)
c=0
for c in range (0,len(pos)):
  pos[c]=pos[c]+20
  c=c+1
```

```
#Sequence Padding
se=seq.values
list1 = []
for s in se:
  s="XXXXXXXXXXXXXXXXXXXX"+s
  s=s+"XXXXXXXXXXXXXXXXXXXX"
  list1.append(s)
  #print(s)
```

```
a = np.zeros(shape=(10000,21))
```

```
f = open("positive.txt", "x")
f = open("positive.txt", "a")
p=0
for ii in list1:
  #print(len(ii),end = '')
  #print("-",end = '')
  for i in ii:
    #print(len(i))
    for j in range(0,len(i)):
        if j==pos[p]:
          #print(j,end='')
          k=j-1
          downstream=k-10
          upstream=k+10
          while downstream<=upstream:
            print(i[downstream],end='')
            f.write(i[downstream])
            downstream=downstream+1          
          #print(p)
          f.write("\n")
          print()
          break
    p=p+1
f.close()
```

```
f = open("negative.txt", "x")
f = open("negative.txt", "a")
for ii in list1:
  for i in ii:
    for j in range(0,len(i)):
       if i[j]=="K":
          downstream=j-10
          upstream=j+10
          while downstream<=upstream:
            print(i[downstream],end='')
            f.write(i[downstream])
            downstream=downstream+1 
          print()
          f.write("\n")       
f.close()
```

```
pos = 'positive.txt'
neg = 'negative.txt'
negative = []
with open(neg, 'r', encoding="utf-8",errors='ignore') as f:
    for line in f:
        negative.append(line.split())
        
positive = []
with open(pos, 'r', encoding="utf-8",errors='ignore') as f:
    for line in f:
        positive.append(line.split())
```

```
for i in negative:
    if i in positive:
        #print "found " + i
        negative.remove(i)
```

```
#convert positive list to array
new_pos_ar=np.array(positive)
pos_uniques = np.unique(new_pos_ar)
len(pos_uniques)
new_neg_ar=np.array(negative)
neg_uniques = np.unique(new_neg_ar)
len(neg_uniques)
pos_uniques_list=pos_uniques.tolist()

neg_uniques_list=neg_uniques.tolist()
```

```
for i in neg_uniques_list:
    if i in pos_uniques_list:
        #print "found " + i
        neg_uniques_list.remove(i)
```

```
with open(' MalonylationPositive.txt', 'w') as f:
    for item in pos_uniques_list:
        f.write("%s\n" % item)
        
with open(' MalonylationNegative.txt', 'w') as f:
    for item in neg_uniques_list:
        f.write("%s\n" % item)
```
